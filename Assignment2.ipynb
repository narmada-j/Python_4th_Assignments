```python
#                                           Python Assignment  2: Data Structures - List, Dictionary, Set & Conditional Statements(Narmada J)
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#        List
# -----------------
# 1. List Creation
# -----------------
import builtins
# a. Create a list named age_list with five integer elements. For eg., [24, 25, 26, 27, 28]
age_list = [24, 25, 26, 27, 28]
print("Age_List : ")
print(age_list)
# b. Create a list named name_list with five string elements.
name_list = ["Arjun", "Priya", "Rahul", "Sneha", "Kathir"]
print("\nName_List : ")
print(name_list)
# ------------------------------------
# 2. List Operations / Modifications:
# ------------------------------------
# a. Append the string "Yazhini" to name_list.
name_list.append("Yazhini")
print("\nAppend : ")
print(name_list)
# b. Insert the element 30 at index 2 in age_list.
age_list.insert(2,30)
print("\nInsert : ")
print(age_list)
# c. Remove the string "Yazhini" from name_list.
name_list.remove("Yazhini")
print("\nRemove : ")
print(name_list)
# d. Pop the last element from age_list.
age_list.pop()
print("\nPop : ")
print(age_list)
# e. Extend the age_list with additional ages [29, 30, 26].
add_age_list = [29,30,26]
age_list.extend(add_age_list)
print("\nExtend : ")
print(age_list)
# f. Sort age_list in descending order.
age_list.sort()
print("\nSort : ")
print(age_list)
# g. Find Max age, Min age and sum of all ages from age_list.
# By Logic
max_age = age_list[0]
min_age = age_list[0]
sum1 = 0
for age in age_list:
  sum1 = sum1 + age
  if(age < min_age):
    min_age = age
  if(age > max_age):
    max_age = age
print("\nLogic \n----------------")
print(f"Min Age : {min_age}")
print(f"Max age : {max_age}")
print(f"Sum Age : {sum1}")
# By method
min1 = builtins.min(age_list)
max1 = builtins.max(age_list)
tot = builtins.sum(age_list)
print("Build  in Function\n-------------------")
print(f"Minimum Age : {min1}")
print(f"Maximum Age : {max1}")
print(f"Total Age   : {tot}")
# -----------------------------
# 3. Accessing List Elements:
# -----------------------------
# a.Print the first element of name_list.
print(f"\nFirst Element : {name_list[0]}")
# b. Print the last element of name_list.
print(f"\nLast Element : {name_list[-1]}")
# c. Print the elements from index 2 to index 4 in name_list.
print(f"\nElement 2 to 4 : {name_list[2:5]}")
# d. Print the elements of name_list in reverse order.
print(f"\nName List : {name_list}")
print(f"\nReverse List : {name_list[::-1]}")
```
