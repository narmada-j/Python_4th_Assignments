```python
#                                                        Python Assignment 1 - Data Structures - Strings & Tuples(Narmada J)
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#          Strings
# ------------------------
# 1. String Concatenation
# ------------------------
string1 = "Hello"
string2 = input("\nEnter Your Name  : ")
string3 = string1 + " " + string2
print(string3)
print(f"\n{string3}, welcome to Python programming\n")
```

```python
# ------------------------------------
#    2. String Slicing and Indexing
#-------------------------------------
inputVal = "Hello zara, welcome to Python programming"
print(f"Input Value is      : {inputVal}")
# a. Print the first character of the string.
print(f"First Character is  : {inputVal[0]}")
# b. Print the last character of the string.
print(f"Last Character is   : {inputVal[-1]}")
# c. Print the first 5 characters of the string.
print(f"First 5 Characters  : {inputVal[:5]}")
# d. Print the last 11 characters of the string.
print(f"Last 11 Characters  : {inputVal[-11:]}")
# e. Print the string in reverse.
print(f"Reverse String      : {inputVal[::-1]}")
# f. Use slicing and print the word “Python” from the existing string.
index = inputVal.find("Python")
sliced = inputVal[index:index+6]
print(f"Slicing Method      : {sliced}\n")
```

```python
# ------------------------------------
#            3. String Methods
#-------------------------------------
strM = "Python beginner tutorial"
# a. Convert the sentence to uppercase.
print(f"Uppercase   : {strM.upper()}")
# b. Convert the sentence to lowercase.
print(f"Uppercase   : {strM.lower()}")
# c. Use Capitalize and return the sentence to the original input form.
print(f"Capitalize  : {strM.capitalize()}")
# d. Count the total number of occurrences of character ‘t’  in the string.
print(f"Count of 't': {strM.count('t')}")
# e. Replace all occurrences of “Python” with “Machine Learning”  in the input string strM = “Python beginner tutorial”
print(f"Replace     : {strM.replace('Python', 'Machine Learning')}\n")
```

```python
# ------------------------------------
#            Tuples
#-------------------------------------
t1 = (10,20,30)
t2 = (40,50,60)
# a. Concatenate the two tuples and store it in “t_combine”
t_combine = t1 + t2
print(f"Combined Tuple:{t_combine}")
# b. Repeat the elements of “t_combine” 3 times
t_repeat = t_combine * 3
print(f"Repeated Tuple: {t_repeat}")
# c. Access the 3rd element from “t_combine”
print(f"Third Element : {t_combine[2]}")
# d. Access the first three elements from “t_combine”
print(f"First Three   : {t_combine[:3]}")
# e. Access the last three elements from “t_combine”
print(f"Last Three    : {t_combine[-3:]}\n")

```
