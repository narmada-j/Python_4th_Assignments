```python
#                                                       Python Assignment 3: While Loop, For loop and Function (Narmada J)
#--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
#       Number Guessing Game
# -----------------------------------
# 1. Set Up the Game:
# ------------------------------------
import random
print("Number Guessing Game\n---------------------")
print("Lets Play...")
# Generate random number
secret_number = random.randint(1,10)
# ------------------------------------
# 2. Prompt the User:
# ------------------------------------
#numb = int(input("Guess Number (1 to 10):"))
attempts = 3
# -------------------------------------------------------
# 3. Implement the Guessing Logic with Control Statements:
# -------------------------------------------------------
attemp = 0
while attemp < attempts:
  numb = int(input("\nGuess Number (1 to 10):"))

  if numb < 1 or numb > 10:
        print("Your Guess is Out of Range...! ")
        attemp += 1
        continue
  if numb > secret_number:
    print("Too High...! Try Again.")
  elif numb < secret_number:
    print("Too Low...! Try Again.")
  elif secret_number == numb:
    print("Congratulations...!You Guess Correct Number...!")
    break
  else:
    print("Invalid...! ")

  attemp += 1
else:
  print(f"\nNo Of Attempt Exceed...!\nBetter luck next time!")
```

```python
# ----------------------------------------------
#                  BMI Calculator
# ----------------------------------------------
def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)
    return bmi
print("\nBMI Calculator \n---------------------")
weight = float(input("\nEnter Your Weight(kg): "))
height = float(input("\nEnter Your Height(Meters): "))
bmi_Val = calculate_bmi(weight,height)
print(f"\nYour BMI is: {bmi_Val:.2f}")
```
